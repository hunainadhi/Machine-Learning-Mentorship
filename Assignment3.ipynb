```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
df = df.iloc[:, 5:]
df
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
y_pred = kmeans.fit_predict(df)
```

```python

plt.figure(figsize=(10, 10))
for i in range(5):
    plt.scatter(df.values[y_pred == i, 0], df.values[y_pred == i, 1],s=50,label = "cluster" + str(i+1))
    
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],s=200, marker='o',c='black',label = "center")
plt.legend();
```

```python
error = []
for k in range(1,15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df.values)
    error.append(kmeans.inertia_)

plt.plot(range(1,15), error)
plt.xlabel("K")
plt.ylabel("Sum of Squares")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
y_pred = kmeans.fit_predict(df)

plt.figure(figsize=(10, 10))
for i in range(3):
    plt.scatter(df.values[y_pred == i, 0], df.values[y_pred == i, 1],s=50,label = "cluster" + str(i+1))
    
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],s=200, marker='o',c='black',label = "center")
plt.legend();
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
dbscan = DBSCAN(eps=0.3, min_samples=10)
```

```python
X = StandardScaler().fit_transform(df.values)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(X)
pca_2d = pca.transform(X)
pca_2d.shape
```

```python
import matplotlib.pyplot as pl
pl.figure(figsize=(20,10))
for i in range(0, 1000):
    if dbscan.labels_[i] == 0:
        c1 = pl.scatter(pca_2d[i,0],pca_2d[i,1],s=100,c='r',marker='^')
    elif dbscan.labels_[i] == 1:
        c2 = pl.scatter(pca_2d[i,0],pca_2d[i,1],s=100,c='g', marker='o')
    elif dbscan.labels_[i] == 2:
        c3 = pl.scatter(pca_2d[i,0],pca_2d[i,1],s=100,c='b', marker='p')
    elif dbscan.labels_[i] == -1:
        c4 = pl.scatter(pca_2d[i,0],pca_2d[i,1],s=100,c='black',marker='*')
pl.legend([c1, c2, c3, c4], ['Cluster 1', 'Cluster 2','Cluster 3','Noise'])
```
