# Covid Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('time_series_2019-ncov-Confirmed.csv')
```

```python
df.head()
```

```python
df=df.groupby(df['Country/Region']).aggregate('sum')
df
```

```python
df.describe()
```

```python
df.columns = pd.to_datetime(df.columns)
```

```python
cols=list(df)
for i in cols:
    x=df[i].value_counts(normalize=True) * 100
    if(x[0]>70):
        df.drop([i],axis=1,inplace=True)
    else:
        break
df
```

```python
plt.plot(df.sum(axis=0));
plt.xticks( rotation='90')
plt.xlabel('Dates')
plt.ylabel('No. of Cases')
plt.show();
```

```python
new_df=df['2020-03-22'].sort_values(ascending=True)
plt.figure(figsize=(27,27))
plt.pie(new_df);

plt.legend(labels=new_df.index)
```
